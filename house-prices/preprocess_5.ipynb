#### import needed libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os 
from sklearn.preprocessing import OneHotEncoder
from scipy import stats
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
```

# Inspect and fix missing values in categorical features

```python
df[cat_cols].isnull().sum()[df[cat_cols].isnull().sum() > 0]
```

### Inspect individual columns if needed

```python
# df["MasVnrType"].value_counts()
```

```python
na_fill_cols = ["Alley", "BsmtQual", "BsmtCond", "BsmtExposure", "BsmtFinType1",
                "BsmtFinType2", "FireplaceQu", "GarageType", "GarageFinish", "GarageQual", 
                "GarageCond", "PoolQC", "Fence", "MiscFeature"]
```

```python
for col in cat_cols:
    if col in na_fill_cols:
        df[col] = df[col].fillna("None")
    else:
        df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df[cat_cols].isnull().sum()[df[cat_cols].isnull().sum() > 0]
```

# Inspect and fix missing values in numerical features

```python
df[num_cols].isnull().sum()[df[num_cols].isnull().sum() > 0]
```

```python
for col in num_cols:
    df[col] = df[col].fillna(0)
df[num_cols].isnull().sum()[df[num_cols].isnull().sum() > 0]
```

# one hot encoding

```python
df_dummies = pd.get_dummies(df)
```

# Add polynomial/other features

```python
df_dummies["age_sold"] = df["YrSold"] - df["YearBuilt"]
df_dummies.loc[df_dummies["age_sold"] < 0, "age_sold"] = 0
df_dummies["age_sold_inv"] = 1 / (df_dummies["age_sold"] + 1)
```

```python
poly_cols = ["LotFrontage", "LotArea", "age_sold", "age_sold_inv", "BsmtFinSF1", "BsmtFinSF2", "BsmtUnfSF", "TotalBsmtSF", "1stFlrSF", "2ndFlrSF", "LowQualFinSF", "GrLivArea", "GarageArea", "WoodDeckSF", "OpenPorchSF", "EnclosedPorch", "3SsnPorch", "ScreenPorch", "PoolArea"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2, include_bias=False)
print("Shape before", df_dummies.shape)
poly_features = poly.fit_transform(df_dummies[poly_cols])
```

```python
poly_df = pd.DataFrame(poly_features, columns=poly.get_feature_names_out(poly_cols))
df_dummies.drop(poly_cols, axis=1, inplace=True)
```

```python
df_dummies.reset_index(inplace=True, drop=True)
poly_df.reset_index(inplace=True, drop=True)
```

```python
df_dummies = pd.concat([df_dummies, poly_df], axis=1)
print("Shape after", df_dummies.shape)
```

# recombine with target and save

```python
target.reset_index(inplace=True, drop=True)
df_dummies = pd.concat([df_dummies, target], axis=1)
df_dummies.head()
```

```python
df_dummies.to_csv(f"data/preprocessed{suffix}.csv", index=False)
```
